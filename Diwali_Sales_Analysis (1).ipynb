**Gender**

```python
df=snr.countplot(x='Gender', data=data)

for bars in df.containers:
  df.bar_label(bars)
snr.set(rc={'figure.figsize':(7,5)})
```

```python
sales_gender= data.groupby(['Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
snr.barplot(x='Gender', y='Amount', data=sales_gender)
snr.set(rc={'figure.figsize':(7,5)})
```

From the above two graph's we are able to know most of the buyers are female & even purchasing power of females are greater than male

**Age**

```python
age=snr.countplot(x='Age Group', hue='Gender', data=data)

for bars in age.containers:
  age.bar_label(bars)
snr.set(rc={'figure.figsize':(10,5)})
#hue for dividing data in to 2 parts like female & Male here
```

```python
age_gender= data.groupby(['Age Group'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
snr.barplot(x='Age Group', y='Amount', data=age_gender)
snr.set(rc={'figure.figsize':(15,5)})
```

From above graphs we understand most buyers age group is 26-35 yrs i.e. female

**State**

```python
Sales_state=data.groupby(['State'],as_index=False) ['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
snr.set(rc={'figure.figsize':(15,5)})
snr.barplot(x='State',y='Orders',data=Sales_state)
#Total numbers of orders from top 10 states
```

```python
Sales_state=data.groupby(['State'],as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
snr.set(rc={'figure.figsize':(15,5)})
snr.barplot(x='State',y='Amount',data=Sales_state)
#Total Amount/Sales of orders from top 10 states
```

from above graphs we can see most of orders & total sales/amount are from Uttar pradesh, Maharastra & Karnataka respectively

**Marital Status**

```python
Married=snr.countplot(x='Marital_Status', data=data)

for bars in Married.containers:
  Married.bar_label(bars)
snr.set(rc={'figure.figsize':(7,5)})
```

```python
married=data.groupby(['Marital_Status','Gender'],as_index=False) ['Amount'].sum().sort_values(by='Marital_Status', ascending=False).head(10)
snr.set(rc={'figure.figsize':(8,5)})
snr.barplot(x='Marital_Status',y='Amount',hue='Gender', data=married)
```

From above the graphs we can see that most of the buyers are married(Women) and they have high purchasing power.

**Occupation**

```python
df=snr.countplot(x='Occupation', data=data)

for bars in df.containers:
  df.bar_label(bars)
snr.set(rc={'figure.figsize':(20,5)})
```

```python
Job_profile=data.groupby(['Occupation'],as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
snr.set(rc={'figure.figsize':(20,5)})
snr.barplot(x='Occupation',y='Amount', data=Job_profile)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare & Aviation sector.

**Product Category**

```python
df=snr.countplot(x='Product_Category', data=data)

for bars in df.containers:
  df.bar_label(bars)
  snr.set(rc={'figure.figsize':(20,5)})
```

```python
Job_profile=data.groupby(['Product_Category'],as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
snr.set(rc={'figure.figsize':(20,5)})
snr.barplot(x='Product_Category',y='Amount', data=Job_profile)
```

From above graphs we can see that most of the sold products are Food, Clothing & Electronics category
